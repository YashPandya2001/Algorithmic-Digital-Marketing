```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url = "https://www.alibaba.com/consumer-electronics/action-sports-camera/p44_p201340102?spm=a2700.8293689.HomeLeftCategory.dlv2category.33d867afPDHSWF"
headers = {"Accept-Language": "en-US, en;q=0.5"}
results = requests.get(url, headers=headers)

soup = BeautifulSoup(results.text, "html.parser")

#initiate data storage
products = []
categories = []
price = []

data_div = soup.find_all('div', class_='lister-item mode-advanced')

#our loop through each container
for container in data_div:

        #product name
        name = container.h3.a.text
        product.append(name)
        
        #price
        year = container.h3.find('span', class_='lister-item-price').text
        years.append(year)

        #categories
        cat_name = container.h3.a.text
        categories.append(cat_name)

#pandas dataframe        
alibaba = pd.DataFrame({
'product': product,
'categories': categories,
'price': price,
})

#cleaning data 
alibaba['product'] = alibaba['product'].str.extract('(\d+)')
alibaba['categories'] = alibaba['categories'].str.extract('(\d+)')
alibaba['price'] = alibaba['price'].astype(int)

#add dataframe to csv file named 'alibaba.csv'
alibaba.to_csv('alibaba.csv')
```
